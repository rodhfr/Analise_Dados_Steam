```python
import pandas as pd

# Carrega o arquivo CSV
df = pd.read_csv('C:/projeto/Analise_Dados_Steam/Analise_Dados_steam/games_data.csv')

# Calcula as médias das colunas 'total reviews', 'positive ratings' e 'negative ratings'
media_total_reviews = df['total reviews'].mean()
media_positiva_reviews = df['positive ratings'].mean()
media_negativa_reviews = df['negative ratings'].mean()

# Calcula as medianas das reviews positivas e negativas
mediana_positiva_reviews = df['positive ratings'].median()
mediana_negativa_reviews = df['negative ratings'].median()

# Calcula a mediana da soma de reviews positivas e negativas
mediana_soma_reviews = (df['positive ratings'] + df['negative ratings']).median()

# Exibe apenas as médias e medianas
print(f"MÉDIAS: ")
print(f'Reviews Totais: {media_total_reviews:.2f}')
print(f'Reviews Positivas: {media_positiva_reviews:.2f}')
print(f'Reviews Negativas: {media_negativa_reviews:.2f}\n')

print(f"MEDIANAS: ")
print(f'Reviews Positivas: {mediana_positiva_reviews:.2f}')
print(f'Reviews Negativas: {mediana_negativa_reviews:.2f}')
print(f'Soma de Reviews Positivas e Negativas: {mediana_soma_reviews:.2f}\n')

# Converte os gêneros em listas de strings, se necessário
df['genres'] = df['genres'].apply(lambda x: eval(x) if isinstance(x, str) else x)

# Expande os gêneros em linhas separadas para cada gênero
df_expanded = df.explode('genres')

# Agrupa por gênero e soma as reviews positivas e negativas
grouped_genres = df_expanded.groupby('genres').agg({
    'positive ratings': 'sum',
    'negative ratings': 'sum'
}).reset_index()

# Encontra o gênero com mais reviews positivas
max_positive_genre = grouped_genres.loc[grouped_genres['positive ratings'].idxmax()]
# Encontra o gênero com mais reviews negativas
max_negative_genre = grouped_genres.loc[grouped_genres['negative ratings'].idxmax()]

# Exibe os gêneros com mais reviews positivas e negativas
print(f'Gênero com mais Reviews Positivas: {max_positive_genre["genres"]} ({max_positive_genre["positive ratings"]})')
print(f'Gênero com mais Reviews Negativas: {max_negative_genre["genres"]} ({max_negative_genre["negative ratings"]})\n')

# Função para garantir que a coluna 'platforms' contenha listas
def ensure_list(x):
    """Função que garante que o valor seja uma lista de strings (plataformas)."""
    if isinstance(x, str):
        try:
            return eval(x)  # Tenta avaliar como lista
        except:
            return [x]  # Caso não consiga, retorna como uma lista de um único item
    elif isinstance(x, list):
        return x  # Já é uma lista
    else:
        return []  # Caso seja nulo ou outro tipo, retorna uma lista vazia

# Aplica a função para garantir que a coluna 'platforms' contenha listas
df['platforms'] = df['platforms'].apply(ensure_list)

# Expande as plataformas em linhas separadas para cada plataforma
df_expanded_platforms = df.explode('platforms')

# Aqui garantimos que o explode funciona corretamente
# Convertemos os valores de 'platforms' para string explicitamente
df_expanded_platforms['platforms'] = df_expanded_platforms['platforms'].astype(str)

# Agrupa por plataforma e soma as reviews positivas e negativas
grouped_platforms = df_expanded_platforms.groupby('platforms').agg({
    'positive ratings': 'sum',
    'negative ratings': 'sum'
}).reset_index()

# Encontra a plataforma com mais reviews positivas
max_positive_platform = grouped_platforms.loc[grouped_platforms['positive ratings'].idxmax()]
# Encontra a plataforma com mais reviews negativas
max_negative_platform = grouped_platforms.loc[grouped_platforms['negative ratings'].idxmax()]

# Exibe as plataformas com mais reviews positivas e negativas
print(f'Plataforma com mais Reviews Positivas: {max_positive_platform["platforms"]} ({max_positive_platform["positive ratings"]})')
print(f'Plataforma com mais Reviews Negativas: {max_negative_platform["platforms"]} ({max_negative_platform["negative ratings"]})')
```
