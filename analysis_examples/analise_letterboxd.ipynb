```python
import pandas as pd

# Ler o arquivo CSV
df = pd.read_csv('diary_nina.csv')

# Verificar se a coluna 'Date' existe
if 'Date' not in df.columns:
    raise ValueError("A coluna 'Date' não foi encontrada no arquivo CSV.")

# Converter a coluna 'Date' para o tipo datetime
df['Date'] = pd.to_datetime(df['Date'])  

# Extrair o dia da semana da coluna de datas
df['dia_da_semana'] = df['Date'].dt.day_name()

# Contar a frequência de cada dia da semana
contagem_dias = df['dia_da_semana'].value_counts()

# Identificar o dia da semana mais frequente
dia_mais_frequente = contagem_dias.idxmax()
frequencia_mais = contagem_dias.max()

# Identificar o dia menos frequente 
dia_menos_frequente = contagem_dias.idxmin()
frequencia_menos = contagem_dias.min()


# Exibir a frequência de todos os dias da semana
for dia, freq in contagem_dias.items():
    print(f'O dia da semana {dia} ocorre {freq} vezes.')

# Calcular a razão entre a frequência do dia mais frequente e o dia menos frequente
razao = frequencia_mais / frequencia_menos

# Exibir os resultados
print(f'O dia da semana mais frequente é {dia_mais_frequente} com {frequencia_mais} ocorrências.')
print(f'O dia da semana menos frequente é {dia_menos_frequente} com {frequencia_menos} ocorrências.')
print(f'Você vê filmes {razao:.1f} vezes mais no dia mais frequente em comparação com o dia menos frequente.')
```

```python
import pandas as pd

# Ler o arquivo CSV
df = pd.read_csv('diary_nina.csv')

# Verificar se as colunas 'Date' e 'Rating' existem
if 'Date' not in df.columns or 'Rating' not in df.columns:
    raise ValueError("As colunas 'Date' e 'Rating' não foram encontradas no arquivo CSV.")

# Converter a coluna 'Date' para o tipo datetime
df['Date'] = pd.to_datetime(df['Date'])

# Converter a coluna 'Rating' para numérico, tratando valores vazios como NaN
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Extrair o dia da semana da coluna de datas
df['dia_da_semana'] = df['Date'].dt.day_name()

# Agrupar por dia da semana e calcular a média das notas
media_notas_por_dia = df.groupby('dia_da_semana')['Rating'].mean()

# Ordenar os resultados para que os dias da semana apareçam na ordem correta
dias_da_semana = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
media_notas_por_dia = media_notas_por_dia.reindex(dias_da_semana)

# Exibir os resultados
for dia, media in media_notas_por_dia.items():
    print(f'A média das notas na {dia} é {media:.2f}.')

# Identificar o dia da semana com a maior e a menor média
dia_maior_media = media_notas_por_dia.idxmax()
maior_media = media_notas_por_dia.max()

dia_menor_media = media_notas_por_dia.idxmin()
menor_media = media_notas_por_dia.min()

# Exibir os resultados
print(f'O dia da semana com a maior média de notas é {dia_maior_media} com {maior_media:.2f}.')
print(f'O dia da semana com a menor média de notas é {dia_menor_media} com {menor_media:.2f}.')
```

```python
import pandas as pd

# Ler o arquivo CSV
df = pd.read_csv('diary_nina.csv')

# Verificar se as colunas 'Date' e 'Rating' existem
if 'Date' not in df.columns or 'Rating' not in df.columns:
    raise ValueError("As colunas 'Date' e 'Rating' não foram encontradas no arquivo CSV.")

# Converter a coluna 'Date' para o tipo datetime
df['Date'] = pd.to_datetime(df['Date'])

# Converter a coluna 'Rating' para numérico, tratando valores vazios como NaN
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Extrair o dia da semana da coluna de datas
df['dia_da_semana'] = df['Date'].dt.day_name()

# Agrupar por dia da semana e calcular a mediana das notas
mediana_notas_por_dia = df.groupby('dia_da_semana')['Rating'].median()

# Ordenar os resultados para que os dias da semana apareçam na ordem correta
dias_da_semana = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
mediana_notas_por_dia = mediana_notas_por_dia.reindex(dias_da_semana)

# Inicializar variáveis temporárias
dia_maior_mediana = None
maior_mediana = -float('inf')

dia_menor_mediana = None
menor_mediana = float('inf')

# Exibir a mediana das notas e encontrar os dias com maior e menor mediana
for dia, mediana in mediana_notas_por_dia.items():
    print(f'A mediana das notas na {dia} é {mediana:.2f}.')
    
    # Atualizar o dia com a maior mediana
    if mediana > maior_mediana:
        maior_mediana = mediana
        dia_maior_mediana = dia
    
    # Atualizar o dia com a menor mediana
    if mediana < menor_mediana:
        menor_mediana = mediana
        dia_menor_mediana = dia

# Exibir os resultados
print(f'O dia da semana com a maior mediana das notas é {dia_maior_mediana} com {maior_mediana:.2f}.')
print(f'O dia da semana com a menor mediana das notas é {dia_menor_mediana} com {menor_mediana:.2f}.')
```
