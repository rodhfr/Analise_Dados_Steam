# Código das Medidas de Centralidade

## Médias e Medianas relativas as vendas dos 1000 primeiros jogos mais vendidos

```python
import pandas as pd
df = pd.read_csv('vgsales.csv')

#1000 primeiros pra análise ficar congruente
df = df.head(1000)

# Desvio Padrao das vendas na América Do Norte, Europa, Japão, outros e Global
std_na = df['NA_Sales'].std()
std_eu = df['EU_Sales'].std()
std_jp = df['JP_Sales'].std()
std_other = df['Other_Sales'].std()
std_global = df['Global_Sales'].std()

# Medianas das vendas na América Do Norte, Europa, Japão, outros e Global
median_na = df['NA_Sales'].median()
median_eu = df['EU_Sales'].median()
median_jp = df['JP_Sales'].median()
median_other = df['Other_Sales'].median()
median_global = df['Global_Sales'].median()

# Medias das vendas na América Do Norte, Europa, Japão, outros e Global
mean_na = df['NA_Sales'].mean()
mean_eu = df['EU_Sales'].mean()
mean_jp = df['JP_Sales'].mean()
mean_other = df['Other_Sales'].mean()
mean_global = df['Global_Sales'].mean()


print(f"Desvio Padrão - NA: {std_na}, EU: {std_eu}, JP: {std_jp}, "
      f"Other: {std_other}, Global: {std_global} \n")

print(f"Mediana - NA: {median_na}, EU: {median_eu}, JP: {median_jp}, "
      f"Other: {median_other}, Global: {median_global} \n")

print(f"Média - NA: {mean_na}, EU: {mean_eu}, JP: {mean_jp}, "
      f"Other: {mean_other}, Global: {mean_global} \n")
```

## Moda Ano que mais aparece na lista e Console que mais aparece

```python
contagem_por_ano = df.groupby('Year').size()

ano_com_mais_jogos = contagem_por_ano.idxmax()
quantidade_jogos_ano = contagem_por_ano.max()

print(f"O ano com o maior número de jogos é {ano_com_mais_jogos} com {quantidade_jogos_ano} jogos.")

contagem_por_console = df.groupby('Platform').size()

console_com_mais_jogos = contagem_por_console.idxmax()
quantidade_jogos_console = contagem_por_console.max()

print(f"O console com o maior número de jogos é {console_com_mais_jogos} com {quantidade_jogos_console} jogos.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar dados
df = pd.read_csv('vgsales.csv')

# Limitando dataframe
df = df.head(100)

# Cálculo das estatísticas
stats = {
    'Desvio Padrão': [df['NA_Sales'].std(), df['EU_Sales'].std(), df['JP_Sales'].std(), df['Other_Sales'].std(), df['Global_Sales'].std()],
    'Mediana': [df['NA_Sales'].median(), df['EU_Sales'].median(), df['JP_Sales'].median(), df['Other_Sales'].median(), df['Global_Sales'].median()],
    'Média': [df['NA_Sales'].mean(), df['EU_Sales'].mean(), df['JP_Sales'].mean(), df['Other_Sales'].mean(), df['Global_Sales'].mean()]
}

# Criar DataFrame para os gráficos
stats_df = pd.DataFrame(stats, index=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales'])

# Plotar
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Gráfico do Desvio Padrão
axs[0].bar(stats_df.index, stats_df['Desvio Padrão'], color='skyblue')
axs[0].set_title('Desvio Padrão das Vendas')
axs[0].set_ylabel('Desvio Padrão')

# Gráfico da Mediana
axs[1].bar(stats_df.index, stats_df['Mediana'], color='lightgreen')
axs[1].set_title('Mediana das Vendas')
axs[1].set_ylabel('Mediana')

# Gráfico da Média
axs[2].bar(stats_df.index, stats_df['Média'], color='salmon')
axs[2].set_title('Média das Vendas')
axs[2].set_ylabel('Média')

plt.tight_layout()
plt.show()
```
